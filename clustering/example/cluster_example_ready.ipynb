```python
import numpy as np
import matplotlib.pyplot as plt             #To plot the clusters
from sklearn.cluster import KMeans          #To perform cluster analysis
from sklearn.datasets import make_blobs     #To create data
from sklearn.metrics import accuracy_score  #To check the accuracy of our model
```

```python
X,y = make_blobs(n_samples=5000,            #How many smaples or the total number of points 
                n_features = 2,             #The numner of features for each sample
                cluster_std = 1,            #Standard deviation of clusters
                shuffle = True)             #Shuffle the data for better distribution
```

```python
X.shape                                     #Check the dimensions of the created test data
```

```python
plt.scatter(X[:,0],X[:,1],c=y)              #Viewing the test data
plt.show()                                  #display data   
```

```python
cluster = KMeans(n_clusters=3)                #Cluster is a variable which holds the values of KMeans
cluster.fit(X)                                #Performing KMeans on the test data X  
centers = cluster.cluster_centers_            #Obtaining the centers the data points or the test data or coordinate of cluster center
labels = cluster.labels_                      #We label each point   
y_pred = KMeans(n_clusters=3).fit_predict(X)  #Predicting the closest cluster each sample in X belongs to
```

```python
plt.scatter(X[:,0],X[:,1],c=y_pred,marker='*')                         #Scattering the test data points 
plt.scatter(centers[:,0],centers[:,1],marker='+',c='r',linewidths=15)  #Scattering the obatinted data coordinates
plt.show()                                                             #Displaying the data  
```

```python
accuracy_score(y,y_pred)                     #viewing the accuracy of our model
```
