```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('diabetic_data.csv')
data.drop('weight',axis=1,inplace=True)
data
```

```python
le = LabelEncoder()
```

```python
std_data = data.apply(LabelEncoder().fit_transform)
std_data.shape
```

```python
clus = KMeans(n_clusters=55,n_jobs=-1)
clus.fit(std_data)
```
