```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('seeds_dataset.txt',header=None,sep=r"\s*")
df = pd.DataFrame(data)
df
```

```python
X = df.as_matrix()
```

```python
clus = KMeans()
clus.fit(X)
```

```python
centroids = clus.cluster_centers_
labels = clus.labels_
centroids,labels
```
